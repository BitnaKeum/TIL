### Instance Normalization

```python
import numpy
# import torch
```

```python
X = np.array([[ 
                 [[1, 2, 9, 2, 7], 
                  [5, 0, 3, 1, 8], 
                  [4, 1, 3, 0, 6], 
                  [2, 5, 2, 9, 5], 
                  [6, 5, 1, 3, 2]], 
                 [[4, 5, 7, 0, 8], 
                  [5, 8, 5, 3, 5], 
                  [4, 2, 1, 6, 5], 
                  [7, 3, 2, 1, 0], 
                  [6, 1, 2, 2, 6]], 
                 [[3, 7, 4, 5, 0], 
                  [5, 4, 6, 8, 9], 
                  [6, 1, 9, 1, 6], 
                  [9, 3, 0, 2, 4], 
                  [1, 2, 5, 5, 2]] 
                 ], 
                [ 
                 [[7, 2, 1, 4, 2], 
                  [5, 4, 6, 5, 0], 
                  [1, 2, 4, 2, 8], 
                  [5, 9, 0, 5, 1], 
                  [7, 6, 2, 4, 6]], 
                 [[5, 4, 2, 5, 7], 
                  [6, 1, 4, 0, 5], 
                  [8, 9, 4, 7, 6], 
                  [4, 5, 5, 6, 7], 
                  [1, 2, 7, 4, 1]], 
                 [[7, 4, 8, 9, 7], 
                  [5, 5, 8, 1, 4], 
                  [3, 2, 2, 5, 2], 
                  [1, 0, 3, 7, 6], 
                  [4, 5, 4, 5, 5]] 
                 ] 
                ]) 
print(X.shape)
```

```python
n, c, h, w = X.shape
```

```python
# 평균
X_mean = np.mean(X.reshape(n, c, -1), axis=2).reshape(n, c, 1, 1)
print(X_mean)
print(X_mean.shape)

### PyTorch
# X_mean = X.view(n, c, -1).mean(dim=2).view(n, c, 1, 1)
```

```python
# 분산, 표준편차
eps = 1e-5
X_var = np.var(X.reshape(n, c, -1), axis=2) + eps
X_std = np.sqrt(X_var).reshape(n, c, 1, 1)
print(X_std)
print(X_std.shape)

### PyTorch
# X_var = X.view(n, c, -1).var(dim=2) + eps
# X_std = X_var.sqrt().view(n, c, 1, 1)
```

```python
# Instance Normalization
output = (X - X_mean) / X_std
print(output)
print(output.shape)
```
