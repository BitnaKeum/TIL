#1. KERAS를 이용한 AND

```
import numpy as np
from keras.models import Sequential
from keras.layers.core import Dense

training_data = np.array([[0,0],[0,1],[1,0],[1,1]], "float32")
target_data = np.array([[0],[0],[0],[1]], "float32")    ### 실제 값

model = Sequential()  ### Sequential : 모델을 만들겠다
#model.add(Dense(1, input_dim=2, activation='sigmoid'))   ### Input layer, Dense : 층을 만드는것, input_dim : 하나의 input은 2개의 값을 가짐
model.add(Dense(16, input_dim=2, activation='relu'))      ### Hidden layer는 1개이고 내부에 노드를 16개 만듦
model.add(Dense(1, activation='sigmoid'))                 ### Output layer

model.compile(loss='mse',
              optimizer='adam',
              metrics=['accuracy'])

model.fit(training_data, target_data, epochs=500, verbose=1)  ### fit : 실제로 running
```

```
model.summary()
```

```
while True: # 두개의 0~1 값을 입력하여 모델 결과를 확인
    inp = list(map(int,input().split()))
    qwe = np.array(inp)
    print("입력된 값 : {}".format(qwe),end="\n\n")
    qwe = qwe.reshape(1,2)
    print("reshape : {}".format(qwe),end="\n\n")
    print("결과값 : {}" .format(model.predict(qwe)[0][0].round()))
    print("")
```

#2. KERAS를 이용한 NAND 문제

```
import numpy as np
from keras.models import Sequential
from keras.layers import Dense

# The input and output, i.e. truth table, of a NAND gate
x_train = np.array([[0,0],[0,1],[1,0],[1,1]], "uint8")
y_train = np.array([[1],[1],[1],[0]], "uint8")

# Create neural networks model
model = Sequential()
# Add layers to the model
model.add(Dense(16, activation='relu', input_dim=2))      # first hidden layer
#model.add(Dense(3, activation='relu'))                   # second hidden layer
model.add(Dense(output_dim=1, activation='sigmoid'))                # output layer

# Compile the neural networks model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
# Train the neural networks model
model.fit(x_train, y_train, nb_epoch=1000)
```

```
while True:
    inp = list(map(int,input().split()))
    qwe = np.array(inp)
    print("입력된 값 : {}".format(qwe),end="\n\n")
    qwe = qwe.reshape(1,2)
    print("reshape : {}".format(qwe),end="\n\n")
    print("결과값 : {}" .format(model.predict(qwe)[0][0].round()))
    print("")
```

#3. KERAS를 이용한 OR 문제

```
import numpy as np
from keras.models import Sequential
from keras.layers.core import Dense

training_data = np.array([[0,0],[0,1],[1,0],[1,1]], "float32")

target_data = np.array([[0],[1],[1],[1]], "float32")

model = Sequential()
model.add(Dense(8, activation='relu', input_dim=2))      # first hidden layer
model.add(Dense(8, activation='relu'))                   # second hidden layer
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='mse',
              optimizer='adam',
              metrics=['accuracy'])

model.fit(training_data, target_data, epochs=500, verbose=1)
```

##4. KERAS를 이용한 XOR 문제 

```
import numpy as np
from keras.models import Sequential
from keras.layers.core import Dense

training_data = np.array([[0,0],[0,1],[1,0],[1,1]], "float32")
target_data = np.array([[0],[1],[1],[0]], "float32")

model = Sequential()
model.add(Dense(4, input_dim=2, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='mse',
              optimizer='adam',
              metrics=['accuracy'])

model.fit(training_data, target_data, nb_epoch=500, verbose=1)
```

```
while True:
    inp = list(map(int,input().split()))
    qwe = np.array(inp)
    print("입력된 값 : {}".format(qwe),end="\n\n")
    qwe = qwe.reshape(1,2)
    print("reshape : {}".format(qwe),end="\n\n")
    print("결과값 : {}" .format(model.predict(qwe)[0][0].round()))
    print("")
```
